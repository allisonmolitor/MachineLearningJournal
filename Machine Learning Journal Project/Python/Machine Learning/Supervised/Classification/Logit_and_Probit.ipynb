Logistic Regression

```python
from google.colab import files

uploaded = files.upload()
```

```python
import pandas as pd
import io
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sea
import statsmodels.tools as sm
import statsmodels.formula.api as smf

dataFrame = pd.read_csv(io.BytesIO(uploaded['Heart_Disease_Prediction.csv']))
dataFrame
```

```python
dataFrame = dataFrame.rename(columns={'Heart Disease': 'HD'})
dataFrame
```

```python

dataFrame['HD'] = dataFrame['HD'].astype('category')
dataFrame['HD'] = dataFrame['HD'].cat.codes
dataFrame
```

```python
log_reg = smf.logit('HD ~ Cholesterol', data=dataFrame).fit()
print(log_reg.summary())
```

```python
pred_prob = np.exp(log_reg.fittedvalues) /( 1+np.exp(log_reg.fittedvalues) )
dataFrame['predict'] = pred_prob
dataFrame
```

Progit Regression

```python
import pandas as pd
import numpy as np
from statsmodels.discrete.discrete_model import Probit
import statsmodels.api as sm
```

```python
Y = dataFrame["HD"]
X = dataFrame["Cholesterol"]
X = sm.add_constant(X)
model = Probit(Y, X.astype(float))
probit_model = model.fit()
print(probit_model.summary())
```

```python
mfx = probit_model.get_margeff(at='mean')
print(mfx.summary())
```
