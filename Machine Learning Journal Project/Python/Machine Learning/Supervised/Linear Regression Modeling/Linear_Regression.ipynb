```python
from google.colab import files

uploaded = files.upload()
```

```python
import pandas as pd
import io
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sea
import statsmodels.tools as sm
import statsmodels.formula.api as smf

dataFrame = pd.read_csv(io.BytesIO(uploaded['raw_data.csv']))
dataFrame
```

```python
dataFrame = dataFrame.dropna()
dataFrame
```

```python
dataFrame['location'] = dataFrame['location'].astype('category')
dataFrame['location'] = dataFrame['location'].cat.codes
dataFrame
```

```python
dataFrame['date'] = dataFrame['date'].astype('category')
dataFrame['date'] = dataFrame['date'].cat.codes
dataFrame
```

```python
x = dataFrame['total_cases']
y = dataFrame['gdp_per_capita']
```

```python
X = dataFrame[[ 'location', 'date', 'total_cases', 'total_deaths', 'stringency_index', 'population', 'human_development_index']]
X
```

```python
from sklearn import linear_model

regr = linear_model.LinearRegression()
regr.fit(X, y)
```

```python
print(regr.coef_)
```

```python
predictedsalary = regr.predict([[153, 290, 8076.0, 230.0, 76.84, 13000000, 0.5]])

print(predictedsalary)
```

```python
OLS_reg = smf.ols('gdp_per_capita ~ total_cases', data=dataFrame).fit()

OLS_reg.summary()
```
